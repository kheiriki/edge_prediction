```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Method': ['Deepwalk', 'Node2vec', 'RGCN', 'RGCN+LSTM', 'RGCN-LSTM (constrain)'],
        'Accuracy': [0.4252, 0.64258, 0.6872, 0.81328, 0.8756],
        'Precision': [0.5123, 0.7614, 0.6512, 0.76232, 0.8123],
        'Recall': [0.5255, 0.7607, 0.70384, 0.85423, 0.79810],
        'AUC': [0.5734, 0.6296, 0.5022, 0.7291, 0.7533]}

df = pd.DataFrame(data)

# Set the 'Method' column as the index of the DataFrame
df.set_index('Method', inplace=True)

# Create a bar chart
fig, ax = plt.subplots()
df.plot(kind='bar', ax=ax, rot=30, color=['Indigo', 'violet', 'pink', 'Plum'])

# Set the title and labels for the chart
ax.set_title('Performance Comparison')
ax.set_xlabel('Method')
ax.set_ylabel('Score')

# Save the plot as a PNG file
fig.savefig('performance_comparison.png')

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

methods = ['Deepwalk', 'Node2vec', 'RGCN', 'RGCN+LSTM', 'RGCN-LSTM (constrain)']
precision = [0.5123, 0.7614, 0.6512, 0.76232, 0.8123]
recall = [0.5255, 0.7607, 0.70384, 0.85423, 0.79810]

# Calculate F1 scores
f1 = [2 * (p * r) / (p + r) for p, r in zip(precision, recall)]

# Plot F1 scores
plt.bar(methods, f1,color= ['purple', 'Indigo', 'violet', 'pink', 'Plum'])
plt.ylim(0, 1)
plt.xlabel('Method')
plt.ylabel('F1 score')
plt.title('F1 scores for Graph-based Models')

# Rotate xlabels and make them colorful
plt.xticks(rotation=45, ha='right')
for i, method in enumerate(methods):
    plt.gca().get_xticklabels()[i]
plt.savefig('f1.png', dpi=300)
# Show the plot
plt.show()

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr = data.corr()

# Plot correlation matrix
sns.set_theme(style="white")
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    cmap = sns.color_palette("Purples", as_cmap=True)
    fig, ax = plt.subplots(figsize=(10, 8))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True, annot=True, fmt='.1f', cmap=cmap)
    ax.set_title('Correlation Matrix of User Features')

# Show the plot
plt.show()
```
